```python
lyric = ["I'm trying to hold my breath",
          "Let it stay this way",
          "Can't let this moment end",
          "You set off a dream with me",
          "Getting louder now",
          "Can you hear it echoing?",
          "Take my hand",
          "Will you share this with me?",
          "'Cause darling without you",
          "All the shine of a thousand spotlights",
          "All the stars we steal from the nightsky",
          "Will never be enough",
          "Never be enough",
          "Towers of gold are still too little",
          "These hands could hold the world but it'll",
          "Never be enough", "For me",
          "Never, never", "Never, for me",
          "Never enough", "For me"]
```

```python
def neverEnough_evo(A): A = [A.insert(i,A[i]) or A for A in [A[:]] for i in [-6,-4,-2,-2,-1,-1,-1] if i in (-1 ,-2 ,-4 ,-6)][0]; f = [w.rstrip(',').rstrip('?') for w in ', '.join(A).split(" ")]; word_count_dic = {word : f.count(word) for word in set(f)}; value_sorted =[word_count_dic[k] for k in sorted(word_count_dic.keys())];print(" --- << Never Enough 2 >> --- ");print(*A, sep='\n');print("\n\n --- <<  Word-Counting the lyric of Never Enough >> --- ");print(*[r.append('{}'.format(sorted(word_count_dic.keys())[i])+ ' '*(20-len(sorted(word_count_dic.keys())[i]) ) + '{}'.format(value_sorted[i])) or r for r in [[]] for i in range(len(value_sorted))][0],sep='\n')
```

```python
neverEnough_evo(lyric)
```
