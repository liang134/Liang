### search function

```python
# define search function
def search(data,y_i=None,y_f=None,teamID=None,playerID=None):  
    if type(y_i)!=int:
        data1=data
    else:data1 = data[data['yearID']>=y_i];
    if type(y_f)!=int:
        data1=data1
    else:data1= data1[data1['yearID']<=y_f];
    if type(playerID)!=str:
        data2=data1
    else:data2 = data1[data1['playerID']==playerID];
    if type(teamID)!=str:
        data3=data2
    else:data3 = data2[data2['teamID']==teamID];
    return data3
```

### choose batter's batting record

```python
# appearance's G_p it means that game as pitcher.
# from appearance choose the player list who is G_p=0
# from the player list find all batting record of them, and save it to batter_bat
batter_bat = pd.DataFrame()
for i in appearance[appearance['G_p']==0].drop_duplicates(subset=['playerID'])['playerID']:
    batter_bat = pd.concat([batter_bat,search(bat,playerID=i)])
```

```python
# to make sure in this data still have pitchers batting record?
for i in batter_bat.merge(appearance[appearance['G_all']==appearance['G_p']][['playerID']]).drop_duplicates(subset=['playerID'])['playerID']:
    batter_bat = batter_bat[batter_bat['playerID']!=i]
```

## Data preprocessing : 
### merge batting record which is splitted

```python
# merge batting record which is splitted 
m = batter_bat.drop_duplicates(subset=['playerID','yearID'])
merge_batter_bat = pd.DataFrame()
for i,j in zip(m['yearID'],m['playerID']):
    f = search(batter_bat,y_i=i,y_f=i,playerID=j)
    if len(f)!=1:
        d = pd.DataFrame(f.sum()).T
        d.loc[:,'playerID']=f['playerID'].drop_duplicates().values
        d.loc[:,'yearID']=f['yearID'].drop_duplicates().values
        d.loc[:,'teamID']=f['teamID'].values[0]
        d.loc[:,'lgID']=f['lgID'].values[0]
        f = d
    merge_batter_bat = pd.concat([merge_batter_bat,f])
```

```python
merge_batter_bat['stint']=1
```

```python
merge_batter_bat = merge_batter_bat.astype({'G':int, 'AB':int, 'R':int, 'H':int,
       '2B':int, '3B':int, 'HR':int, 'RBI':int, 'SB':int, 'CS':int, 'BB':int, 'SO':int, 'IBB':int, 'HBP':int, 'SH':int,
       'SF':int, 'GIDP':int})
```

```python
merge_batter_bat
```

```python
mbb = merge_batter_bat.copy()
mbb = mbb[mbb['lgID']!=0]
```

```python
mbb
```

```python
# cumulation batting table
cbt = pd.DataFrame()
for i in mbb.drop_duplicates(subset=['playerID'])['playerID']:
    g = search(mbb,playerID=i)
    k = np.cumsum(g)
    k['playerID']=i
    k['yearID']=g['yearID']
    k['teamID']=g['teamID']
    k['lgID']=g['lgID']
    cbt = pd.concat([cbt,k])
cbt['age']=0
```

```python
cbt.describe(include='all').T
```

```python
# add seniority but not rename
mbb['stint'] = cbt['stint']
```

```python
# rename seniority
mbb = mbb.rename(columns={'stint':'seniority'})
```

```python
# batting record located on 1990-2020 regular season
mbb_1990 = mbb[mbb['yearID']>=1990]
mbb_1990['age'] = 0
```

```python
# adding player's age
for i in range(len(mbb_1990)):
    mbb_1990.iloc[i,len(mbb_1990.columns)-1] = int(mbb_1990.iloc[i,1])-int(search(People,playerID=mbb_1990.iloc[i,0]).iloc[0,1])
```
